# Experimenting with Polars instead of Pandas

```python
import polars as pl
```

```python
df = pl.read_csv('./data/bmi.csv')
df.shape

```

```python
df.describe()
```

```python
print(df.sample(fraction=0.1))
```

```python
# New columns called bmi
df.with_columns(
    pl.struct(pl.col(['Height', 'Weight']))
        .apply(lambda x:  x['Weight'] / ((x['Height']/100) ** 2))
        .alias('bmi')
)
```

```python
df.to_dummies(["Gender"])
```
